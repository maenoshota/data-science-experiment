````python
# 準備
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# データの読み込み 
df = pd.read_csv("2023年mlb.csv", sep=",", encoding="utf-8")
# 別途、前処理が必要であればすること
# データの内容を確認
df
````

````python
# データの傾向把握
# 分析に使用する変数のみを選択して、各種の統計量を確認
df[["2塁打", "HR"]].describe()
````

````python
# 目的変数を取得
y = df["2塁打"]
# 説明変数を取得
x = df["HR"]
# 目的変数と説明変数の散布図を描画
plt.scatter(x, y)
````

````python
# 相関分析の実施
# 相関分析の有意水準を設定
p_s = 0.05
# 相関係数とp値を取得
r, p = pearsonr(x,y)
# 相関分析の結果を表示
if p < p_s:
    print(f"x と y には相関があり、相関係数は{r}であった。")
elif p >= p_s:
    print(f"x と y には相関がない可能性を排除できなかった。")
````

### 相関分析

相関分析は目的変数と説明変数の間の**相関の強さ**、つまり**説明変数によって目的変数の値がどれほど決定付けられるか**が興味の対象である。  
一方、回帰分析は**説明変数が増えると目的変数がどれほど増えるか**が興味の対象である。

相関分析で求める**相関係数**(ピアソンの積率相関係数)の式  
$$\begin{align*}
r_{xy} &= \frac{\sum_{i=1}^{n}\left(x_i - \bar{x}\right)\left(y_i - \bar{y}\right)}{\sqrt{\sum_{i=1}^n\left(x_i-\bar{x}\right)^2}\sqrt{\sum_{i=1}^n\left(y_i-\bar{y}\right)^2}}\\
&= \frac{\frac{1}{n}\sum_{i=1}^{n}\left(x_i - \bar{x}\right)\left(y_i - \bar{y}\right)}{\sqrt{\frac{1}{n}\sum_{i=1}^n\left(x_i-\bar{x}\right)^2}\sqrt{\frac{1}{n}\sum_{i=1}^n\left(y_i-\bar{y}\right)^2}}\\
&= \frac{共分散_{xy}}{標準偏差_x 標準偏差_y}
\end{align*}$$  

相関分析で行う検定 : **無相関の検定**  
$H_0$ : 母相関係数は0である(2つの変数には相関がない)  
$H_1$ : 母相関係数は0ではない(2つの変数には相関がある)  
t分布に従う検定統計量 : $t = \frac{|r|\sqrt{n-2}}{\sqrt{1-r^2}}$

相関分析で**できること、できないこと**

|できる|できない|
|:---|:---|
|相関係数を知る<br/>相関の有無を判断する|一方の変数の値からもう一方の変数の値を推定する|


### データの傾向把握を行う際のチェックリスト

* **目的変数**と**説明変数**は**連続値**であるか
* **目的変数**は**どのような分布**であるか
  * 分布の形状は不自然でないか
  * 負の数はあるか
  * ゼロの値は取りうるか
  * 平均値はいくつか
  * 標準偏差は平均値に比べてどの程度か
  * 外れ値はあるか
* **説明変数**は**どのような分布**であるか
  * 分布の形状は不自然でないか
  * 負の数はあるか
  * ゼロの値は取りうるか
  * 平均値はいくつか
  * 標準偏差は平均値に比べてどの程度か
  * 外れ値はあるか
* 目的変数と説明変数、**それぞれの変数の関係は線形**か
  * 非線形ならばどのような変数変換をすれば線形となるか

### データが回帰分析に適さないケース

* 目的変数や説明変数単体の特徴
  * **名義尺度**や**順序尺度**である
  * 2〜3種類の**少ない種類の値**しか取っていない
* 目的変数と説明変数の関係
  * 一見して**直線的な関係になっていない**
  * 極端な **外れ値（直線から大きく外れるデータ点）** が存在する
  * 1つの説明変数で**目的変数の取りうる値が複数ある**

````python
import statsmodels.api as sm
````

````python
# 切片を推定するために定数列を追加
X = sm.add_constant(x) # 左辺は大文字
# 回帰モデル（最小二乗法モデル）を定義
model = sm.OLS(y, X)
# モデルの推定
res = model.fit()
````

````python
# 切片と傾きの取得
beta0_est, beta1_est = res.params

# モデル式の描画
# モデル式による目的変数の推定結果を取得
y_est = beta0_est + beta1_est*x
# 散布図とモデルの推定結果を描画
plt.scatter(x, y, alpha=0.5)
plt.plot(x,y_est,c="r")
plt.show()

# 誤差の計算
resid = y_est - y

# 誤差の可視化
plt.scatter(x, resid, alpha=0.5)
plt.show()

# 誤差の2乗和の評価
np.sum(resid**2)
````

解釈

2塁打に対する本塁打の単回帰係数の推定値は0.9145であった。  
すなわち、本塁打が1本増加すると2塁打は0.9145本プラスになる。  
これは、HRバッターはボールを遠くに飛ばす技術に優れているため2塁打も出やすいのだと考えられる。 

また、切片の推定値は2.6912であった。  
すなわち、本塁打が0である時の2塁打は2.6912本であると推定された。  
これは、本塁打を狙うよりもヒットを狙うほうが簡単だからだと思われる。

ただし、上記の議論は本塁打が0本から35本の範囲でのみ有効である。  
また、2塁打と本塁打の関係性は必ずしも因果を説明している訳ではない。  
すなわち、2塁打を増やすために本塁打を操作する事が意味をなさない場合がある。  
また、この分析では本塁打の他に2塁打と関係を持つ要因を一切考慮していない。  
すなわち、データに加えられていない何らかの要因によって、本塁打と2塁打に擬似的な相関関係が生じている可能性を否定できない。

````python
# QQプロットの描画関数をインポート
from scipy.stats import probplot
# QQプロットの描画範囲を設定
ax = plt.subplot()
# QQプロットを描画
probplot(resid, plot=ax)
plt.show()
````

残差のヒストグラムを見ると左右対称であり、1つの山を持つつりがね型の形状をするため、正規分布に近い形状をしている事が分かる。  
残差プロットを見ると目的変数の推定値は残差に依存しているため、独立であると言うことはできない。  
QQプロットを見るとデータ点が直線上に概ね載っているため、正規分布に近い形状をしている事が分かる。  
以上の事から、今回の回帰分析モデルは誤差が独立な正規分布から発生しているという前提を満たすと判断される。

````python
# MSEとRMSEの計算
from sklearn.metrics import mean_squared_error
import numpy as np

y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]
mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
````
